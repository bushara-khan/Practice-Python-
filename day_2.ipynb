Operators in python 

1. Arithmetic Operators 

```python
a = 9
b = 2

print(a % b)  # Output: 0
#power 
print(a ** b)
#floor division
print(a // b)  # Output: 4
```

2. Assingment operators

```python
x = 5
y = 10

y += 5 # y = y + 5 
print(y)  # Output: 15

x *= 2 # x = x * 2
print(x)
```

3. Comparison (Relational Operators )

```python
a = 10
b = 3 

print( a == b )
print( a != b )
print( a > b )
print(a < b)
print( a >= b )
print(a <= b)
```

```python
4. # Logical Operators and, or, not
# and -> multiplication -> 0 - > false ,1 -> true 
# or -> addition 
# not -> negation

p = 10
q = 4 

print ( (p > 5) and (q < 3) )  # True and false  -> false
print ( (p > 5) or (q < 3) )   # True or false  -> true
print ( not(p > 5) )            # not true -> false
```

5. Membership operatos in, not in

```python
# Membership Operators

fruits = ["apple", "banana", "cherry"]
```

```python
print("xyz" not in fruits)  # Output: true
```
