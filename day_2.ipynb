# Operators in python 

 ## 1. Arithmetic Operators 
 Arithmetic operators in Python are used to perform basic mathematical operations on numbers.
 some arithmetic operators:


```python
a = 9
b = 2

print(a + b)  # Output: 11 addition
print(a - b)  # Output: 7 subtraction
print(a * b)  # Output: 18 multiplication
print(a / b)  # Output: 4.5 division
print(a % b)  # Output: 0 reminder 
#power 
print(a ** b)
#floor division
print(a // b)  # Output: 4
```

## 2. Assingment operators
Assignment operators in Python are used to assign values to variables.
They can also update the value of a variable.

#### Basic assignment


```python
x = 5
y = 10

```

other assignment operators

```python
y += 5 # y = y + 5  Addition Assignment Operator (+=)
print(y)  # Output: 15

x-=2 # x = x - 2  Subtraction Assignment Operator (-=)
print(x)  # Output: 3 

x *= 2 # x = x * 2  Multiplication Assignment Operator (*=)
print(x)
y /= 5 # y = y / 5
print(y)  # Output: 3.0 Division Assignment Operator (/=)
```

## 3. Comparison (Relational Operators )

Comparison operators are used to compare two values.
They always return True or False.

```python
a = 11
b = 3 

print( a == b ) # Equal to
print( a != b ) # Not equal to
print( a > b ) # Greater than
print(a < b) # Less than
print( a >= b ) # Greater than or equal to
print(a <= b) # Less than or equal to
```

Logical operators are used to combine conditions.

```python
4. # Logical Operators and, or, not
# and -> multiplication -> 0 - > false ,1 -> true 
# or -> addition 
# not -> negation

p = 10
q = 4 

print ( (p > 5) and (q < 3) )  # True and false  -> false
print ( (p > 5) or (q < 3) )   # True or false  -> true
print ( not(p > 5) )            # not true -> false
```

## 5. Membership operatos in, not in 
Membership operators are used to check whether a value exists in a sequence
like a list, tuple, or string.

There are two membership operators:
in
not in

```python
# Membership Operators

fruits = [ "banana", "cherry"]
```

```python
print("apple"  in fruits)  
print("banana" in fruits)
print("grape" not in fruits)
```
