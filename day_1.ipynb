# Variable
A variable is a temporary storage in which some value is stored for further use.
### Syntax:
Syntax to create variable is: 

variable_name = value

```python
# i store student name in std_n variable.
std_n =  "john david"
student_roll_number = 2
```

can also be initialized in one line

```python
x,y,z = 1,2,3
```

### Rules for variable naming:
1. A variable name must start with a letter or underscore
2. It cannot start with a number
3. Spaces are not allowed in variable name
4. Special symbols ($, &,@, etc) are not allowed in variable name
5. Variables are case sensitive means "abc" and "ABC" are different

```python
name = "Ali"
Name = "Ahmed"
```
