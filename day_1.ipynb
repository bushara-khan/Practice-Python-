# Variables
Variables are temporary storage used to store data in a program. They act as containers for values that can be changed or manipulated during the execution of a program.
### Syntax:
variable_name = value

### Example:
age = 18
name = "Ayaan"
* (Here 'age' is a variable holding a numeric value, and 'name' is a variable holding text)

```python
# Storing student data
std_n = "john david"
student_roll_number = 2

print(std_n)
print(student_roll_number)
```

### Rules for variable naming:
1. A variable name must start with a letter or underscore
2. It cannot start with a number
3. Spaces are not allowed in variable name
4. Special symbols ($, &, @, etc) are not allowed
5. Variables are case sensitive (e.g., "abc" and "ABC" are different)

# Datatypes
In Python, data types define the kind of value a variable can hold.

1. **Integer (int):** Whole numbers (positive, negative, or zero) without decimals.
2. **Float:** Numbers with a decimal point.
3. **String (str):** Text wrapped in quotes.
4. **Boolean (bool):** True or False values.

```python
# Examples of datatypes
student_age = 20          # Integer
student_gpa = 3.5         # Float
student_name = "Ali Raza" # String
is_passed = True          # Boolean
```
